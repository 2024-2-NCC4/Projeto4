```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Lista de todos os aspectos possíveis e seus novos nomes
aspectos_possiveis = [
    'Senioridade das vagas recebidas em relação à sua experiência',
    'Aprovação em processos seletivos/entrevistas',
    'Oportunidades de progressão de carreira',
    'Velocidade de progressão de carreira',
    'Nível de cobrança no trabalho / Stress no trabalho',
    'Atenção dada pelas pessoas diante das minhas opiniões e ideias',
    'Relação com outros membros da empresa em momentos de integração e outros momentos fora do trabalho',
    'Relação com outros membros da empresa em momentos de trabalho'
]

# Novo nome para cada aspecto
novos_nomes_aspectos = [
    'Vagas vs Experiência',
    'Aprovação em Entrevistas',
    'Oportunidades de Progressão',
    'Velocidade de Progressão',
    'Cobrança e Stress no Trabalho',
    'Atenção às Ideias',
    'Relação Fora do Trabalho',
    'Relação no Trabalho'
]

# Criar um dicionário de mapeamento entre o nome original e o novo nome
mapa_renomeacao = dict(zip(aspectos_possiveis, novos_nomes_aspectos))

# Contar quantas vezes cada aspecto aparece no campo
def contar_aspectos(texto, aspectos_possiveis):
    if pd.notnull(texto):
        count = {aspecto: texto.count(aspecto) for aspecto in aspectos_possiveis}
        return count
    return {aspecto: 0 for aspecto in aspectos_possiveis}

# Supondo que seu DataFrame tenha uma coluna chamada 'dsc_P1f_aspectos_prejudicados'
# que contém as descrições dos aspectos prejudicados

# Aplicando a contagem de aspectos no campo relevante
contagem_total = df_dados_2023['dsc_P1f_aspectos_prejudicados'].apply(contar_aspectos, args=(aspectos_possiveis,))

# Agregando as contagens de cada aspecto
contagem_aspectos = pd.DataFrame(list(contagem_total)).sum()

# Filtrando apenas os aspectos com contagem maior que zero
contagem_aspectos = contagem_aspectos[contagem_aspectos > 0]

# Ordenando os aspectos pela frequência
sorted_indices = np.argsort(contagem_aspectos)[::-1]
contagem_aspectos = contagem_aspectos.iloc[sorted_indices]
aspect_labels = contagem_aspectos.index

# Renomeando os aspectos usando o mapa de renomeação
aspect_labels_renomeados = [mapa_renomeacao[aspecto] for aspecto in aspect_labels]

# Plotando o gráfico
plt.figure(figsize=(10, 6))

# Usando uma paleta de cores
cmap = plt.cm.get_cmap('plasma')
colors = cmap(np.linspace(0, 1, len(contagem_aspectos)))

# Criando o gráfico de barras
plt.bar(aspect_labels_renomeados, contagem_aspectos, color=colors[:len(contagem_aspectos)])

# Ajustando a exibição
plt.xticks(rotation=45, ha='right')
plt.ylabel('Contagem')
plt.title('Aspectos Afetados pela Diversidade no Profissional')

plt.tight_layout()
plt.show()
```

```python
# DISTRIBUIÇÃO POR GÊNERO
import matplotlib.pyplot as plt

def distribuicao_genero():
  contagem_genero = df_dados_2023['dsc_genero'].value_counts()
  contagem_genero = contagem_genero.groupby(lambda x: x if x not in ['Prefiro não informar', 'Outro'] else 'Prefiro não dizer / Outros').sum()
  cmap = plt.cm.get_cmap('plasma')
  colors = cmap(np.linspace(0, 1, len(contagem_genero)))
  plt.figure(figsize=(7, 7))
  plt.pie(contagem_genero, labels=None, autopct='%1.1f%%', startangle=140, colors=colors, pctdistance=1.2)
  plt.legend(contagem_genero.index, title="Gênero", loc="center left", bbox_to_anchor=(1.2, 0.5))
  plt.title('Distribuição por Gênero')
  plt.tight_layout()
  plt.show()

if __name__ == "__main__":
  distribuicao_genero()
```

```python
# DITRIBUIÇÃO POR FAIXA SALARIAL (GRÁFICO DE BARRAA)
import matplotlib.pyplot as plt
import numpy as np

def faixa_salarial_barras():
    # Contanto frequência
    num_faixa_salarial = df_dados_2023['dsc_faixa_salarial'].value_counts()
    num_faixa_salarial = num_faixa_salarial[num_faixa_salarial > 0]
    num_faixa_salarial = num_faixa_salarial.sort_values(ascending=False)  # ordenando decrescente
    aspect_labels = num_faixa_salarial.index # (índices)
    # Gráfico
    plt.figure(figsize=(10, 6))
    cmap = plt.cm.get_cmap('plasma')
    colors = cmap(np.linspace(0, 1, len(num_faixa_salarial)))
    plt.bar(aspect_labels, num_faixa_salarial, color=colors[:len(num_faixa_salarial)])
    plt.xticks(rotation=45, ha='right')
    plt.ylabel('Contagem')
    plt.title('Distribuição das Faixas Salariais')
    plt.tight_layout()
    plt.show()

# Testando
if __name__ == "__main__":
    faixa_salarial_barras()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def regiao():
  # Dicionário mapeando os estados para suas respectivas regiões
  mapa_regioes = {
      'São Paulo (Sp)': 'Sudeste', 'Alagoas (Al)': 'Nordeste', 'Distrito Federal (Df)': 'Centro-Oeste',
      'Bahia (Ba)': 'Nordeste', 'Ceará (Ce)': 'Nordeste', 'Paraná (Pr)': 'Sul',
      'Rio Grande Do Sul (Rs)': 'Sul', 'Espírito Santo (Es)': 'Sudeste', 'Paraíba (Pb)': 'Nordeste',
      'Minas Gerais (Mg)': 'Sudeste', 'Santa Catarina (Sc)': 'Sul', 'Goiás (Go)': 'Centro-Oeste',
      'Rio De Janeiro (Rj)': 'Sudeste', 'Rondônia (Ro)': 'Norte', 'Piauí (Pi)': 'Nordeste',
      'Mato Grosso (Mt)': 'Centro-Oeste', 'Pará (Pa)': 'Norte', 'Rio Grande Do Norte (Rn)': 'Nordeste',
      'Pernambuco (Pe)': 'Nordeste', 'Amazonas (Am)': 'Norte', 'Sergipe (Se)': 'Nordeste',
      'Mato Grosso Do Sul (Ms)': 'Centro-Oeste', 'Tocantins (To)': 'Norte', 'Acre (Ac)': 'Norte',
      'Amapá (Ap)': 'Norte', 'Maranhão (Ma)': 'Nordeste'
  }

  # Verificar se 'dsc_estado' existe no seu DataFrame e se a coluna não tem valores nulos
  if 'dsc_estado' in df_dados_2023.columns:
      df_dados_2023['dsc_estado'] = df_dados_2023['dsc_estado'].str.title()
      df_dados_2023['regiao'] = df_dados_2023['dsc_estado'].map(mapa_regioes)
      df_dados_2023['regiao'].fillna('Desconhecida', inplace=True)
      # Contar a frequência das regiões
      contagem_regioes = df_dados_2023['regiao'].value_counts()
      # Criando o gráfico de pizza
      plt.figure(figsize=(7, 7))
      cmap = plt.cm.get_cmap('plasma')
      colors = cmap(np.linspace(0, 1, len(contagem_regioes)))
      plt.pie(contagem_regioes, labels=None, autopct='%1.1f%%', startangle=140, colors=colors, pctdistance=1.2)
      # Adicionar a legenda
      plt.legend(contagem_regioes.index, title="Região", loc="center left", bbox_to_anchor=(1,  0.5))
      # Título e ajuste
      plt.title('Distribuição Regional no Brasil')
      plt.tight_layout()
      plt.show()
  else:
      print("A coluna 'dsc_estado' não foi encontrada no DataFrame.")

if __name__ == "__main__":
  regiao()
```
